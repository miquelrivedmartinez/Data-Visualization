```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_excel('EUR_USD Historical Data.xlsx')

fig = go.Figure(data=go.Ohlc(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Price']))

fig.update_layout(
    title='EUR - USD 2021'
)

fig.show()
```

```python
import matplotlib.pyplot as plt

df_1 = pd.read_excel('evolparo.xlsx')

plt.figure(num=3, figsize= (15,7))
plt.plot(df_1['Mes'], df_1['Paro'], color = "green",
        linewidth=1.0, linestyle = '-')
plt.title('Evolucin del paro 2020')
plt.ylabel('Paro en Millones')
plt.xlabel('Meses')
```
